```python
# put all the lables in a list
labels = ['Anger', 'Fear', 'Joy', 'Sadness', 'Surprise']

# check if the labels is balanced.
for label in labels:
  print(train_data[label].value_counts(normalize = True))
```

- Anger: This label is higly imbalanced as '0' class has a greater proportion.
- Fear: This is relatively balanced, as the class proportion does not differ so much.
- Joy: It is imbalanced.
- Sadness: Moderately Imbalanced.
- Suprise: Moderately Imbalanced.

> Most of the labels are imbalanced, especially Anger and Joy. This imbalance can affect model performance, especially if the model has a bias towards predicting the majority class.
